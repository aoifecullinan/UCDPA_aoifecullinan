```python
## group regions by the mean of the bmi, age and charges  
df_region = insurance_file.groupby('region',as_index=False).agg({'bmi':'mean','age':'mean', 'charges': 'mean'})
print(df_region)
```

```python
## subplot of average bmi, age, charges in relation to the 4 regions 
fig, ax = plt.subplots(2,2, figsize = (10,10))
sns.set_style('darkgrid')
sns.set_palette('colorblind')
sns.set_context('notebook')
fig1= sns.barplot(x='region', y='bmi',data=df_region, ax=ax[0,0])
fig2= sns.barplot(x='region', y='age',data=df_region, ax=ax[0,1])
fig3= sns.barplot(x='region', y='charges',data=df_region, ax=ax[1,0])
fig.delaxes(ax[1][1])


plt.xticks(rotation=70)
## set all y labels
fig1.set_ylabel('Average Bmi')
fig2.set_ylabel('Average Age')
fig3.set_ylabel('Average Charges')

## set all x labels 
fig1.set_xlabel('Regions')
fig2.set_xlabel('Regions')
fig3.set_xlabel('Regions')


#set title 
fig.suptitle('BMI, Age, And Charges in Relation to Regions', y=1)

```
